```python
import requests
import xml.etree.ElementTree as ET
import os
from dotenv import load_dotenv
load_dotenv()

url = 'https://apis.data.go.kr/B553881/newRegistlnfoService_02/getnewRegistlnfoService02'


service_key = os.getenv('PUBLIC_SERVICE_KEY')

# url = 'http://apis.data.go.kr/B553881/newRegistInfoService/newRegistInfoService'                 


params = {
    'serviceKey' : service_key,
    'registYy' : '2026',
    'registMt' : '02',
    'vhctyAsortCode' : 1,
    'registGrcCode' : 1
}

requests.get(url,params=params)

response = requests.get(url,params=params)
print(f'코드 : {response.status_code}')
# print(response.text) 문자열 xml 구조
# print(type(response.text))

root = ET.fromstring(response.text)

# root.findtext('header/resultCode')
# root.findtext('header/resultMsg')
# root.findtext('body/dtaCo')

for child in root.find('header'):  # Element 객체
    print(child.tag, child.text)

for child in root.find('body'): # Element 객체
    print(child.tag, child.text)

print(root.findtext('body/dtaCo'))
```
